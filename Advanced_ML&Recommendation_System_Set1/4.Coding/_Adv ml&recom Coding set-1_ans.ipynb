```python
# Code
import pandas as pd
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
df = pd.DataFrame({
    'Age': [25, 32, 47, 51, 62],
    'Salary': [40000, 50000, 60000, 52000, 62000],
    'Purchased': [0, 1, 1, 0, 1]
})
X = df[['Age', 'Salary']]
y = df['Purchased']
test = SelectKBest(score_func=f_regression, k=1)
fit1=test.fit(X,y)
features_name=X.columns[test.get_support()]
print(list(features_name))
```

```python
# Code
import pandas as pd

data = pd.DataFrame({
    'User': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Item': ['Apple', 'Banana', 'Apple', 'Orange', 'Banana', 'Orange'],
    'Rating': [5, 3, 4, 2, 5, 4]
})
# The below code take the mean rating for each item and sort the top 2 values based on the mean rating 
top2_items = (data.groupby('Item')['Rating'].mean().reset_index().sort_values(by='Rating',ascending=False).head(2))
print(list(top2_items['Item']))
```

```python
from sklearn.datasets import load_diabetes
import pandas as pd
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

X, y = load_diabetes(return_X_y=True, as_frame=True)
lin = LinearRegression()
rfe=RFE(estimator=lin, n_features_to_select=5)
rfe.fit(X,y)
selected_features = X.columns[rfe.support_]
print("Selected features:", list(selected_features))
```

```python
#4: Code
from sklearn.datasets import load_digits
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import mutual_info_classif
import pandas as pd
X,y = load_digits(return_X_y=True, as_frame=True)
test = SelectKBest(score_func=mutual_info_classif, k=8)
X_new = test.fit_transform(X,y)

selected_features = X.columns[test.get_support()]
print("Top 8 features using mutual info:", list(selected_features))


```

```python
#5.
from sklearn.datasets import load_breast_cancer
import pandas as pd

X, y = load_breast_cancer(return_X_y=True, as_frame=True)
Write a Python program to perform feature selection using SelectFromModel. Train a Logistic Regression model and select features based on importance using SelectFromModel. Print the names of the selected feature


Expected Output:
Selected features: ['mean radius', 'mean concavity', 'texture error', 'worst texture', 'worst smoothness', 'worst compactness', 'worst concavity', 'worst concave points', 'worst symmetry']
```

```python
#5.Code
from sklearn.datasets import load_breast_cancer
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectFromModel
X, y = load_breast_cancer(return_X_y=True, as_frame=True)

selector = SelectFromModel(LogisticRegression(
        max_iter=10000))
selector.fit(X,y)


selected_features = X.columns[selector.get_support()]
print("Selected features:", list(selected_features))
```
