### 5. Working on location column

```python
zomato_chn_dataset["Location"].value_counts()
```

```python
def clean_value(val):
    # If there's a comma, return the part after it
    if ',' in val:
        return val.split(',')[1].strip()  # strip() removes leading/trailing spaces
    # Otherwise, return the value unchanged
    return val
zomato_chn_dataset['Location'] = zomato_chn_dataset['Location'].apply(clean_value)
```

```python
zomato_chn_dataset["Location"].value_counts()
```

```python
zomato_chn_dataset_check=zomato_chn_dataset[zomato_chn_dataset["Name of Restaurant"]=="Cocktail Peninsula"]
zomato_chn_dataset_check
```

```python
# There are 5 records with the t.nagar area without space after the dot
zomato_chn_dataset_location_value = zomato_chn_dataset[zomato_chn_dataset["Location"]=="T.Nagar"]
zomato_chn_dataset_location_value
```

```python
# Replacing them with correct values
zomato_chn_dataset.replace({"Location":"T.Nagar"},"T. Nagar", inplace=True)
```

```python
zomato_chn_dataset["Location"].value_counts()
```

```python
# Replacing the hotel name available in Location with correct location name(Nungamvakkam-the slate)
zomato_chn_dataset.replace({"Location":"The Slate"},"Nungambakkam", inplace=True)
```

```python
zomato_chn_dataset["Location"].value_counts()
```

```python
# Finding the duplicate records
zomato_chn_dataset[zomato_chn_dataset.duplicated()]
```

### Find the list of Affordable Restaurants?

#### The criteria for Affordable Restaurants would be:-

#### 1) Low Price 2)High Rated

```python
#### First step will be to find the restaurants with average cost 1/4th the average cost of most expensive restaurant in our dataframe.
#### The most expensive restaurant has an average meal cost= 5000. We'll try to stay economical and only pick the restaurants that are 1/4th of 5000.
```

```python
max_price = zomato_chn_df['Price for 2'].max()
max_price
```

```python
one_fourth_price = max_price/4
one_fourth_price
```

```python
# Finding list of restaurants that have price less than and equal to 1/4th of the max price i.e Finding Cheap Restaurant
aff_rest_df=zomato_chn_df[["Name of Restaurant","Price for 2","Cuisine","Location"]]
aff_rest_df=aff_rest_df[aff_rest_df["Price for 2"]<=1250]
aff_rest_df.sort_values(by="Price for 2", inplace=True)
aff_rest_df
```

```python
# Finding the highest rated list of restaurants

highrate_rest_df= zomato_chn_df[['Name of Restaurant','Price for 2','Cuisine','Location','Delivery Rating']]
highrate_rest_df=highrate_rest_df[highrate_rest_df['Delivery Rating']>=4.5]
highrate_rest_df.sort_values(by='Price for 2',inplace=True)
highrate_rest_df
```

#### Now, we'll merge the aff_rest_df with highrate_rest_df to obtain the intersection i.e the list of Affordable Restaurants !!

```python
highrate_aff_df=pd.merge(aff_rest_df, highrate_rest_df, how='inner', on=['Name of Restaurant', 'Location'])
highrate_aff_df=highrate_aff_df[['Name of Restaurant', 'Price for 2_x','Cuisine_x', 'Location']]                        
highrate_aff_df.rename(columns={'Name of Restaurant':'Name of Restaurant','Price for 2_x' : 'Price for 2',
                                'Cuisine_x':'Cuisine','Location': 'Location'}, inplace=True)
```

```python
highrate_aff_df
```

### Find the list of most Reliable Restaurants?

#### The criteria for most Reliable Restaurants would be:-

#### 1) Low Price 2) High Rated 3) Higher delivery count
#### First step will be to find the restaurants with greater delivery count

```python
mean_count = zomato_chn_df['Delivery Rating Count'].mean()
mean_count
```

```python
# Finding list of restaurants that have delivery Rating Count greater than and equal to Delivery Rating Count
mean_rest_df = zomato_chn_df[['Name of Restaurant', 'Price for 2', 'Cuisine', 'Location', 'Delivery Rating Count']]
mean_rest_df = mean_rest_df[mean_rest_df['Delivery Rating Count'] > 713]
mean_rest_df.sort_values(by='Delivery Rating Count', inplace=True)
mean_rest_df
```

## These are the most reliable, highest rated and affordable restaurants:-

### We obtain this dataframe by simply taking the intersection of highrate_aff_df & mean_rest_df
### This dataframe obtained below shows the restaurants whose:

* Cost is below 1250
* Rating is above 4.5
* Delivery rating count are above 713

```python
reliable_rest_df = pd.merge(mean_rest_df, highrate_aff_df, how='inner', on=['Name of Restaurant', 'Location'])
reliable_rest_df = reliable_rest_df[['Name of Restaurant','Price for 2_x', 'Cuisine_x', 'Location']]
reliable_rest_df.rename(columns={'Name of Restaurant':'Name of Restaurant', 'Price for 2_x':'Price for 2', 'Cuisine_x':'Cuisine',
'Location':'Location'},inplace=True)
```
