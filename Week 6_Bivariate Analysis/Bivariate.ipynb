```python
# separating quan and qual columns from the dataset
quan=[]
qual=[]
def quanqual(dataset):
    for columnName in dataset.columns:
        if(dataset[columnName].dtype=="object"):
            qual.append(columnName)
        else:
            quan.append(columnName)
    return quan,qual
```

```python
quan,qual = quanqual(dataset)
print(quan,"\n",qual)
```

```python
# to remove multicolinearity using Variance Inflation factor
# importing variance_inflation_factor from outliers_influence of stats of statsmodel module
from statsmodels.stats.outliers_influence import variance_inflation_factor
# function named as calc_vif and quantitative columns of dataset is passed as parameter
def calc_vif(X):
# creating a dataframe named vif with variable values respect to the dataset columns passed to the function
   vif=pd.DataFrame()
   vif["variables"] = X.columns
# for each value available in column, variance inflation factor is calculated and stored in the VIF variable of vif table(DataFrame)
   vif["VIF"] = [variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
   return(vif)
```

```python
calc_vif(dataset[quan])
```

```python
# removing degree pass mark
calc_vif(dataset[["ssc_p","hsc_p","etest_p","mba_p","salary"]])
```
