```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
delays=[15, 30, 45, 20, 25, 100, 5, 60, 35, 50, 
          120, 80, 10, 75, 90, 200, 55, 40, 70, 150]

flight_delays_df = pd.DataFrame(delays, columns=["Delay"])

# Percentiles
percentile_10th = flight_delays_df["Delay"].quantile(0.10)
percentile_25th = flight_delays_df["Delay"].quantile(0.25)
percentile_50th = flight_delays_df["Delay"].quantile(0.50)
percentile_75th = flight_delays_df["Delay"].quantile(0.75)
percentile_90th = flight_delays_df["Delay"].quantile(0.90)

print(f"Percentiles (10th, 25th, 50th, 75th, 90th): [{percentile_10th} {percentile_25th} {percentile_50th} {percentile_75th} {int(percentile_90th)}]")

# IQR and bounds
IQR = percentile_75th - percentile_25th
lower_bound = percentile_25th - (1.5 * IQR)
upper_bound = percentile_75th + (1.5 * IQR)
print("IQR:", IQR)
lower_bound=percentile_25th - (1.5*IQR)
upper_bound=percentile_75th + (1.5*IQR)
#print(lower_bound, upper_bound)
outliers = []
for delay in flight_delays_df["Delay"]:
    if delay < lower_bound or delay > upper_bound:
        outliers.append(delay)
print("Outliers:", outliers)
sns.boxplot(flight_delays_df, x="Delay")
plt.xlabel("Flight delays(minutes)")
plt.title("Box plot for flight delays")
plt.show()

```

```python
Question 2 :
A company wants to analyze the salary distribution of its employees to understand the central tendency and determine whether the data is skewed.
DataSet:
salaries = [30000, 32000, 35000, 37000, 40000, 42000, 45000, 47000, 50000, 55000,
            60000, 62000, 65000, 67000, 70000, 72000, 75000, 80000, 85000, 90000]
Expected Output:
Mean Salary: 56950.0
Median Salary: 57500.0
Mode Salary: 30000
The data is Left Skewed (Negative Skew)
histogram
```

```python

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
salaries = [30000, 32000, 35000, 37000, 40000, 42000, 45000, 47000, 50000, 55000,
            60000, 62000, 65000, 67000, 70000, 72000, 75000, 80000, 85000, 90000]
salary_dataset = pd.DataFrame(salaries, columns=["Salary"])
mean_salary = salary_dataset["Salary"].mean()
median_salary = salary_dataset["Salary"].median()
mode_salary=salary_dataset["Salary"].mode()[0]
print("Mean Salary: ", mean_salary)
print("Median Salary: ", median_salary)
print("Mode Salary: ", mode_salary)
if(mean_salary>median_salary):
    print("The data is right skewed(Positive skew)")
elif(mean_salary<median_salary):
    print("The data is Left skewed(Negative Skew)")
else:
    print("The data is symmetrical(zero skew)")
sns.histplot(data=salary_dataset["Salary"], kde=True)
plt.title("Histogram of Employee Salaries")
plt.show()
```

```python
Question 5 : A company conducted a customer satisfaction survey where customers rated their experience on a scale of 1 to 10. Analyze the survey results to calculate descriptive statistics and visualize the distribution of customer satisfaction ratings.
Sample DataSet:  
ratings = [8, 9, 7, 5, 6, 10, 9, 4, 7, 8, 
           6, 9, 10, 5, 8, 7, 6, 9, 10, 7]
Expected Output:
Mean Rating: 7.5
Median Rating: 7.5
Mode Rating: 7
Standard Deviation: 1.746424919657298
```

```python
import numpy as np
from scipy import stats
ratings= [8, 9, 7, 5, 6, 10, 9, 4, 7, 8, 
           6, 9, 10, 5, 8, 7, 6, 9, 10, 7]
mean_rating = np.mean(ratings)
print("Mean Rating:", mean_rating)
median_rating=np.median(ratings)
print("Mediain Rating:",median_rating)
mode_rating=stats.mode(ratings)[0]
print("Mode Rating: " , mode_rating)
standard_deviation=np.std(ratings)
print("Standard Deviation: ", standard_deviation)
```
