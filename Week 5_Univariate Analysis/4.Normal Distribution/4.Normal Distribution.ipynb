# Using for loop to generate values between range and their probabilites

```python
# function to display the probability density function for the given range
def get_pdf_probability(dataset,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax=sns.distplot(dataset,kde=True,kde_kws={'color':"red"},color='Green')
    pyplot.axvline(startrange,color='blue')
    pyplot.axvline(endrange,color='blue')
    # generate a sample
    sample=dataset
    # calculate the mean and standard deviation
    sample_mean=sample.mean()
    sample_std=sample.std()
    print('Mean=%.3f,Standard deviation=%.3f'%(sample_mean,sample_std))
    # define the distribution
    dist=norm(sample_mean,sample_std)

    # sample probabilities for a range of outcomes
    values=[]
    for value in range(startrange,endrange):
        values.append(value)
    print(values)
    probabilities=[]
    for value in values:
        probabilities.append(float(dist.pdf(value)))
    print("\n",(probabilities))
    prob=float(sum(probabilities))
    print("the area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob
 # replacing the 3 line for loop code with single inliner also works in the same way    
```

```python
get_pdf_probability(dataset["ssc_p"],50,80)
```

# Explaining code line by line for standard normal distribution

```python
# Converting normal distribution to Standard normal distribution
# defining the function with function name stdNBgraph and dataset to be passed as parameter
def stdNBgraph(dataset):
# import seaborn library for plotting the graph    
    import seaborn as sns
# calculating the mean for the given column in the dataset using mean() function and assign it to mean variable
    mean=dataset.mean()
# calculating the standard deviation for the given column in the dataset using std() function and assigne it to std variable
    std=dataset.std()
# each values of particular column in dataset is store in list i and assigned to  list values    
    values = [i for i in dataset]
# z_score is calculated for each values in dataset and stored to list z_score    
    z_score =[((j-mean)/std) for j in values]
# density plot with respect to the given column in the dataset is plotted using distplot method by passing z_score as parameters
# kde=True so the curve is drawn in the plot    
    sns.distplot(z_score,kde=True)
# summation of the z_score divided by number of z_score values is calculated    
    (sum(z_score)/len(z_score))
```

```python
stdNBgraph(dataset["ssc_p"])
```
