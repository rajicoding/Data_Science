```python
# to display quantitative and qualitative columns in a separate list
# create an empty list for quan and qual
quan=[]
qual=[]
# Check columns one by one using for loop
for columnName in dataset.columns:
    if(dataset[columnName].dtype=="O"):
        qual.append(columnName)
    else:
        quan.append(columnName)
print(quan)
print(qual)
```

```python
# having the above code in a single function by passing the dataset as parameter
def quanqual(dataset):
    quan=[]
    qual=[]
    # Check columns one by one using for loop
    for columnName in dataset.columns:
        if(dataset[columnName].dtype=="O"):
            qual.append(columnName)
        else:
            quan.append(columnName)
    return quan,qual
```

```python
quan,qual= quanqual(dataset)
```

```python
print(quan)
print(qual)
```

```python
# to calculate percentile for placement dataset
# define the table rows and columns
descriptive=pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%","Q3:75%","Q4:100%"],columns=quan)
descriptive
```

```python
descriptive=pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%","Q3:75%","99%","Q4:100%"],columns=quan)
# using for loop to update required values for each column value from quan
for columnName in quan:
     descriptive.loc["Mean",columnName] = dataset[columnName].mean()
     descriptive.loc["Median",columnName] = dataset[columnName].median() 
     descriptive.loc["Mode",columnName] = dataset[columnName].mode()[0]
     descriptive.loc["Q1:25%",columnName] = dataset.describe()[columnName]["25%"]
     descriptive.loc["Q2:50%",columnName] = dataset.describe()[columnName]["50%"]
     descriptive.loc["Q3:75%",columnName] =  dataset.describe()[columnName]["75%"]
     descriptive.loc["Q4:100%",columnName] =  dataset.describe()[columnName]["max"]
     descriptive.loc["99%",columnName] =  np.percentile(dataset[columnName],99)
```
