```python
# to display quantitative and qualitative columns in a separate list
# create an empty list for quan and qual
quan=[]
qual=[]
# Check columns one by one using for loop
for columnName in dataset.columns:
    if(dataset[columnName].dtype=="O"):
        qual.append(columnName)
    else:
        quan.append(columnName)
print(quan)
print(qual)
```

```python
# having the above code in a single function by passing the dataset as parameter
def quanqual(dataset):
    quan=[]
    qual=[]
    # Check columns one by one using for loop
    for columnName in dataset.columns:
        if(dataset[columnName].dtype=="O"):
            qual.append(columnName)
        else:
            quan.append(columnName)
    return quan,qual
```

```python
quan,qual= quanqual(dataset)
```
