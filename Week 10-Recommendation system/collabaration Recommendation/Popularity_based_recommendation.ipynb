```python
# importing libraries
import pandas as pd
import numpy as np
```

```python
#Read the movie Dataset
i_cols = ['movie id', 'movie title' ,'release date','video release date', 'IMDb URL', 'unknown', 'Action', 'Adventure',
    'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
    'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
items = pd.read_csv('ml-100k/u.item', sep='|', names=i_cols,encoding='latin-1')
```

```python
# Select only needed Item
movie_content=items[['movie id','movie title','Action', 'Adventure',
    'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
    'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']]
```

```python
#read the ratings dataset
r_cols = ['user_id', 'movie id', 'rating', 'unix_timestamp']
ratings = pd.read_csv('ml-100k/u.data', sep='\t', names=r_cols,encoding='latin-1')
```

```python
category_wise_rating = pd.merge(movie_content,ratings, on='movie id', how='inner')
category_wise_rating
```

```python
# To calculate average rating per genre
avg_action=category_wise_rating.loc[category_wise_rating['Action']==1,'rating'].mean()
```

```python
print("Action:", round(avg_action, 2))
```

```python
# To calculate user count per genre
user_count_action = category_wise_rating.loc[category_wise_rating['Action']==1,'user_id'].nunique()
```

```python
print("user_count_Action:", user_count_action)
```

```python
# Function to identify top5_movies for the given df and genre
def top5_movies(df,column):
    top5_movies =(df[df[column] == 1]
              .groupby('movie title')['rating']
              .mean()
              .reset_index() 
              .sort_values(by='rating', ascending=False) 
              .head(5) ) 
    return top5_movies
```

```python
top5_movies(category_wise_rating,'Action')
```

```python
# for calculating top5 in all genres
def top5_all_genres(df):
    genres = ['Action', 'Adventure',
    'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
    'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
    # to store top 5 movies per genre
    top5_movie = {}
    for g in genres:
        # to check the flag as 1, whether it is watched or not
        genre_df=df[df[g]==1]
        # Average ratings if the same movie is watched by multiple users 
        agg = genre_df.groupby('movie title')['rating'].mean().reset_index()
        # top5 movies based on ratings by sorting them
        top5_movie[g] = agg.sort_values(by='rating',ascending=False).head(5)
    # to display result for each genre
    # Here we had defined the top5_movie as dictionary
    # key(g) is the genre and value(movies) is the top 5 movies for that particular genre 
    # items() is the dictionary method that returns pairs of key and value:
    for g, movies in top5_movie.items():
        print(f"\n Top 5 {g} movies:")
        print(movies)
  return(top5_movie)  
```

```python
top5_all_genres(category_wise_rating)
```
